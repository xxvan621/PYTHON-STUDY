```python
def average_score(scores):
    total=0
    for score in scores:
        total+=score
    return total/len(scores)
print(average_score([12,34,56,78,90]))
```

```python
def sum(n):
    total=0
    for i in range(0,n+1):
        total+=i
    return total
print(sum(100))
```

```python
def max(setting):
    max_value=setting[0]
    for i in setting:
        if i>setting[0]:
            max_value=i
    return max_value
pp=[1,23,4,56,67,8,99]
print(max(pp))
```
